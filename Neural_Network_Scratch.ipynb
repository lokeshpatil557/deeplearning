```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
df=pd.read_csv("insurance_data.csv")
df.head()
```

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest = train_test_split(df.iloc[:,:-1],df.iloc[:,-1],test_size=0.3,random_state=1)

xtrain_scale = xtrain.copy()

xtrain_scale["age"]=(xtrain_scale["age"]-xtrain_scale["age"].min())/(xtrain_scale["age"].max()-xtrain_scale["age"].min())

xtest_scale = xtest.copy()

xtest_scale["age"]=(xtest_scale["age"]-xtest_scale["age"].min())/(xtest_scale["age"].max()-xtest_scale["age"].min())
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential([ Dense(1,input_shape=(2,),activation="sigmoid",kernel_initializer="ones",bias_initializer="zeros") ] )

model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])

model.fit(xtrain_scale,ytrain,epochs=500)
```

```python
model.evaluate(xtest,ytest)
```

```python
import numpy as np
import math

class NeuralNetwork:

  def __init__(self):
    self.w1=1.0
    self.w2=1.0
    self.bias=0
  
  def fit(self,x,y,epochs,loss_threshold):
    self.w1,self.w2,self.bias=self.gradient_descent(x['age'],x['afford'],y,loss_threshold,epochs)
    return None

  def sigmoid_function(self,wightsum):
    return 1/(1+np.exp(-wightsum))

  def predict(self,age,afford):
    weightedsum = self.age*self.w1 + self.afford*self.w2 + self.bias
    return self.sigmoid_function(weightedsum)

  def log_loss(self,ytrue,ypred):
    ep=1e-15
    ypred_new =[max(i,ep) for i in ypred]
    ypred_new =[min(i,1-ep)for i in ypred_new]
    ypred_new=np.array(ypred_new)
    return -np.mean(ytrue * np.log(ypred_new) + (1-ytrue) *(1-np.log(ypred_new)))

  def gradient_descent(self,age,afford,ytrue,loss_threshold,epochs):
    
    w1=w2=1
    bias=0
    rate=0.5
    n=len(age)
    for i in range(epochs):
      sumOfWeight = w1*age+w2*afford+bias
      ypredict=self.sigmoid_function(sumOfWeight)
 
      loss=-self.log_loss(ytrue,ypredict)
      
      w1d=1/n * np.dot(np.transpose(age),(ypredict-ytrue))
      w2d=1/n * np.dot(np.transpose(afford),(ypredict-ytrue))
      basisd=np.mean(ypredict-ytrue)

      w1=w1-rate*w1d
      w2=w2-rate*w2d
      bias=bias-rate*basisd

      print(f'epoch : {i} w1 : {w1} w2 : {w2} bias: {bias} loss: {loss}')
      if loss<=loss_threshold:
        break
    return w1,w2,bias

```
