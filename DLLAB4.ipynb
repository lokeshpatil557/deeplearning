```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow import keras
import tensorflow as tf
from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
```

```python
a=pd.DataFrame(dataset['data'],columns=dataset['feature_names'])
a.head()
```

```python
a['feature_names']=dataset['target']
a.head()
```

```python
from sklearn.model_selection import train_test_split
X=a.drop('feature_names',axis=1)
Y=a['feature_names']
xtrain,xtest,ytrain ,ytest  = train_test_split(X,Y,test_size=0.30,random_state=1)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(10,input_shape=(4,),activation='relu'))
```

```python
#from keras.backend import spare_categorical_crossenyropy
model.compile(
    optimizer='adam',
    loss ='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(xtrain,ytrain,epochs=200)
```
